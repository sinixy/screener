```python
import pandas as pd
import yfinance
from tradingview_screener import Scanner, Query, Column
from datetime import datetime
from time import sleep
```

```python
nrows, df = (Query()
 .select('close', 'volume', 'market_cap_basic', 'premarket_change')
 .where(
     Column('type') == 'stock',
     Column('is_primary') == True,
     Column('market_cap_basic') > 30e6,
     Column('premarket_change') > 3
 )
 .order_by('premarket_change', ascending=False)
 .get_scanner_data()
)
df
```

```python
for i, row in df.iterrows():
    ticker = yfinance.Ticker(row.ticker)
    print(f'================== {row.ticker} - {row.premarket_change :.2f}% ==================')
    for news in ticker.get_news():
        dt = datetime.fromtimestamp(news["providerPublishTime"])
        print('! ' * (dt.date() == datetime.today().date()) + f'[{dt}]', news['title'])
    print()
```
